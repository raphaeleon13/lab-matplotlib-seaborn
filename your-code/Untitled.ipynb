```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.arange(0,10,1)
x2 = np.linspace(0,5,10)
plt.plot(x1, x1, label='linear')
plt.plot(x1, x2**2, label='quadratic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Linear vs Non Linear Plots")
plt.legend()
plt.show()
```
